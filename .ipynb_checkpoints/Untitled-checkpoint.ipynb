```python
import pandas as pd
import numpy as np

import datetime as dt

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://data.ontario.ca/dataset/f4112442-bdc8-45d2-be3c-12efae72fb27/resource/455fd63b-603d-4608-8216-7d8647f43350/download/conposcovidloc.csv'

df = pd.read_csv(url)

df.head()
```

## Filter the data set for 'Not Resolved' cases, after June 1 and under 20 year olds

```python
df['Test_Reported_Date'] = pd.to_datetime(df['Test_Reported_Date'])
```

```python

start_date = '2020-06-01'
df = df[(df['Test_Reported_Date'] >= start_date) & (df['Outcome1'] == "Not Resolved") & (df['Age_Group'] == '<20')]
```

### Keep only required rows for analysis 

```python
min(df.Test_Reported_Date)
```

```python
cols = ['Test_Reported_Date','postal_code_first_three', 'Case']
df2 = df.loc[:, cols]
df2.head()
```

```python
datetime_index = pd.DatetimeIndex(df2.Test_Reported_Date.values)
df3 = df2.set_index(datetime_index)
df3.drop('Test_Reported_Date', axis = 1, inplace = True)
df3 = df3.sort_index()
df3.head()
```

```python
df4 = df3.groupby(['postal_code_first_three'])['Case'].sum()
df5 = pd.DataFrame(df4)
df5.plot.bar()
```

## Using apply to create a new column from postal codes to extract the first 3 letters

```python
def get_element(my_list, position):
    return my_list[position]
```

```python
df["postal_code_first_three"] = df["Reporting_PHU_Postal_Code"].str.split(" ").apply(get_element, position=0)

# this can be with a lambda function
# df["postal_code_first_three"] = df["Reporting_PHU_Postal_Code"].str.split(" ").apply(lambda x: x[0])
```
